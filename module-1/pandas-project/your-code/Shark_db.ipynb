```python
import numpy as np

import pandas as pd
data = pd.read_csv('GSAF5.csv', encoding = "ISO-8859-1", delimiter = ',')
data.head()
```

```python
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

```python
data.shape
```

```python
data.drop(columns=["Unnamed: 22"], axis=1, inplace=True)
data = data.drop(columns=["Unnamed: 23"], axis=1)
data.shape
```

```python
drop_cols = list(null_cols[null_cols > 10000].index)
data = data.drop(drop_cols, axis=1)
data.head()
```
